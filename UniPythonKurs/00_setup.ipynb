### Setup `nbstripout`
To avoid diff noise and terrible merge conflicts, we use [nbstripout](https://github.com/kynan/nbstripout) as a commit filter: only input cells will be tracked by git, but not outputs/execution counts/... . This happens automatically with the installation steps below.
Beware though: if you want to pull/merge changes, you should manually run `nbstripout` on your notebooks before.

# Git usage

To use git and make use of `nbstripout` from the terminal (either in JupyterHub or using ssh), use the proper python installation by typing
```
module load python/3.9.2-infiniband
```

Before running a `git pull` or `git merge`, it is recommended to manually run `nbstripout` on all notebook files, so any actual merge conflicts of *input cells* can be handled properly without the noise of the *output cells*:
```
nbstripout *.ipynb
```

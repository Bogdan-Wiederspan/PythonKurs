## Conway's Game of Life
[Conway's Game of Life (GoL)](https://en.wikipedia.org/wiki/Conway%27s_Game_of_Life) is a cellular automaton that consists of cells in a two dimensional array that are either dead or alive.
For the next generation, there is a set of rules to determine whether a cell will be dead or alive:
- Any live cell with fewer than two live neighbours dies, as if by underpopulation.
- Any live cell with two or three live neighbours lives on to the next generation.
- Any live cell with more than three live neighbours dies, as if by overpopulation.
- Any dead cell with exactly three live neighbours becomes a live cell, as if by reproduction.

Based on these rules, there are special starting patterns (= configurations of alive cells) that e.g. result in recurring configurations.
An example is the blinker:
```bash
./gol.py --insert-blinker 4 5
```



An example of a moving pattern is the glider:
```bash
./gol.py --insert-glider 2 3 --update-interval .1
```




If you start with a random configuration, eventually you will end up in a completely static state:
```bash
./gol.py --init-random .3
```

## Gas simulation

We want to simulate a gas.
A gas consists of gas particles (atoms or molecules), which either attract each other or push away from each other due to physical forces.
Whether they attract each other or push away from each other depends on the distance of the particles to each other.
At very small distances they push away from each other.
This is described by the Lennard Jones potential.
Taking care of correctly implementing all the forces between the particles is not possible in this course.
So we make some assumptions and simplifications, which basically reduces our problem ton an ideal gas or a billard table:
- Particles move with constant velocity.
- There are no forces between the particles.
- If particles come very close to each other (distance is smaller than some predefined threshold) we have an elastic collision (elastischer Sto) between those particles.
- We simulate small time steps of movement and check afterwards whether there are collisions.
- We model a 2D volume.
- We can't simulate infinite large volumes, such that we restrict ourselfs to a box with length $L$ with periodic boundary conditions.
  If a particle leaves the box at the right border, it enters again from the left.
- Every particle has a mass.

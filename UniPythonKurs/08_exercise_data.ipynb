### Problem 1

Write a function `sum_file` that takes the single parameter `filename` of type string as argument.
The function should read the file `filename` that contains integer numbers separated by line breaks and return the sum of all these numbers.
The file can contain empty lines, which thus need to be ignored. If the file does not contain any integer at all, zero should be returned.

You do not need to consider floats or any other data type.
A line will always contain a single integer or be empty, like in this example:

````python
%%writefile exercise_data_numbers.txt
23
42
2
34

-556
65
345

345
5
53

-35
6
85698

324

24
````

````python
# PROBLEM (2)

# SOLUTION
def sum_file(filename):
    with open(filename) as file:
        lines = file.read().splitlines()
        #lines = file.readlines()
        print(repr(lines))
    mysum = 0
    for n in lines:
        if not n:
            continue
        mysum += int(n)
    return mysum

# PROBLEM-TEST
with open("mycrypticfilename.txt", "w") as file:
    file.writelines("\n".join(map(str, [1, 2, 3])))
s1 = sum_file("mycrypticfilename.txt")
with open("mycrypticfilename.txt", "w") as file:
    file.writelines("\n\n\n")
s2 = sum_file("mycrypticfilename.txt")
with open("mycrypticfilename.txt", "w") as file:
    file.writelines("\n".join(map(str, [234, 234, 3, -123, "", "", 55, 999, 0, "", 30])))
s3 = sum_file("mycrypticfilename.txt")

int(s1), int(s2), int(s3)
````

### Problem 3

Write a function that:
- Is called `write_calc` 
- Has three parameters: the string `filename`, the float `value1` and the float `value2`
- Calculates the sum, difference, and product of the two numbers `value1` and `value2`
- Stores the results in the file with the name passed by the `filename` parameter
- Each of the three results should be written to its own line and the order should be: sum, difference, product

````python
# PROBLEM (2)

# SOLUTION

def write_calc(filename="", value1=0.0, value2=0.0):
    the_sum = value1 + value2
    the_diff = value1 - value2
    the_prod = value1 * value2
    with open(filename, "w") as myfile:
        myfile.write(str(the_sum)+"\n")
        myfile.write(str(the_diff)+"\n")
        myfile.write(str(the_prod)+"\n")

# PROBLEM-TEST
write_calc(filename="exercise_8_3.txt", value1=137.035, value2=1.602)
with open("exercise_8_3.txt", "r") as myfile:
    formatted_lines = [round(float(v),1) for v in myfile.read().strip("\n").splitlines()]
formatted_lines
````

#### Problem 5

Write a function that:
- Is called `calc_row_mean` 
- Has one parameter: A nested list of the measurements (format of the output of task 4)
- Calculates the mean of each row. For example, the mean for the row `10.3,4.5,3.3,1.7` is `(10.3 + 4.5 + 3.3 + 1.7) / 4.0`.
- Rounds the means to one decimal. You can use the `round(value, 1)` function for this 
- Return a nested list that contains the existing table and an additional column with the name `Mean` that contains the means of each row. So if the input table looks like

```
[[Meas1,Meas2,Meas3,Meas4],
 [10.3,4.5,3.3,1.7],
 [4.2,6.5,4.3,2.2]]
```
the output table should look like
```
[[Meas1,Meas2,Meas3,Meas4,Mean],
 [10.3,4.5,3.3,1.7,4.95],
 [4.2,6.5,4.3,2.2,4.3]]
```

*Hint: You could use the `sum(list)` function that sums the elements of a list. The `len(list)` function might also be useful.*


````python
# PROBLEM (2)

# SOLUTION

def calc_row_mean(inputs):
    # list to hold the outputs until we write them
    outputs = []
    # now lets loop over the rows
    row_counter = 0
    for row in inputs:
        # first row is the header
        if row_counter == 0:
            # add new column called Mean
            outputs.append(row + ["Mean"])
        else:
            # we need the values in float instead of string
            # so we loop over each value and transform it into a float which we then store in the float_list
            float_list = []
            for val in row:
                float_list.append(float(val))
            # the sum
            the_sum = sum(float_list)
            # the mean
            the_mean = the_sum / len(float_list)
            the_mean = round(the_mean, 1)
            # now we append the input values of the row together with the str(the_mean) to the outputs list
            outputs.append(row + [str(the_mean)])
        # increase the counter
        row_counter += 1
    return outputs

# PROBLEM-TEST
measurement_list = [
    ["Meas1", "Meas2", "Meas3", "Meas4"],
    ["1.3", "2.2", "7.5", "9.1"],
    ["2.6", "8.3", "5.9", "5.1"],
    ["1.1", "0.8", "9.4", "3.5"],
    ["5.5", "8.3", "2.7", "9.5"],
    ["1.8", "2.9", "9.4", "2.0"],
    ["4.5", "5.7", "4.3", "0.7"],
    ["6.9", "4.5", "1.5", "7.3"],
    ["7.6", "6.9", "7.7", "1.7"],
    ["0.7", "9.3", "8.2", "5.4"],
    ["9.6", "9.9", "4.1", "4.6"],
]
[[str(val) for val in row] for row in calc_row_mean(measurement_list)]
````

#### Problem 6

Write a function `write_experiment_csv` with two parameters.
The first parameter gives the name of the file where the output is written to and the second parameter is a nested list (in the format of the output of task 5) that contains the measurements and the mean values.
The function should write the content of the nested list to the given **csv** file.

````python
# PROBLEM (1)
import csv  # use the csv module to parse the file contents

# SOLUTION

def write_experiment_csv(output_filename, outputs):
    with open(output_filename, "w") as myfile:
        csv_writer = csv.writer(myfile)
        for row in outputs:
            csv_writer.writerow(row)

# PROBLEM-TEST
measurement_list = [
    ["Meas1", "Meas2", "Meas3", "Meas4", "Mean"],
    ["1.3", "2.2", "7.5", "9.1", "5.0"],
    ["2.6", "8.3", "5.9", "5.1", "5.5"],
    ["1.1", "0.8", "9.4", "3.5", "3.7"],
    ["5.5", "8.3", "2.7", "9.5", "6.5"],
    ["1.8", "2.9", "9.4", "2.0", "4.0"],
    ["4.5", "5.7", "4.3", "0.7", "3.8"],
    ["6.9", "4.5", "1.5", "7.3", "5.0"],
    ["7.6", "6.9", "7.7", "1.7", "6.0"],
    ["0.7", "9.3", "8.2", "5.4", "5.9"],
    ["9.6", "9.9", "4.1", "4.6", "7.1"],
]
write_experiment_csv("exercise_8_4_output_exam.csv", measurement_list)

# check output
with open("exercise_8_4_output_exam.csv", "r") as myfile:
    content = myfile.read()
content.strip().lower()
````

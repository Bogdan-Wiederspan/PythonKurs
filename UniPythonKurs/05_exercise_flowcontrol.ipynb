### Problem 2

Write a function `is_prime` that checks whether a number (argument of the function) is prime or not.

```python
# PROBLEM (1)

# SOLUTION
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False
    return True

# PROBLEM-TEST
is_prime(739397), is_prime(739398), is_prime(8887), is_prime(7693)
```

### Problem 4

An object is moving on a predefined path (e.g. train on tracks, a marble on a marble run).
Along the path are several photoelectric sensors (Lichtschranken).
Any time the object passes a photoelectric sensor, a measurement device records the current time and the position of the sensor.
Unfortunately these photoelectric sensors are not very reliable and sometimes a sensor is not registering the passage of the object.

Write a function `velocity_calculator(positions, times)` that calculates for each section of the path the average velocity of the object.
Each section starts at a photoelectric sensor (except the last one) and ends at the next one (i.e. the first section starts at the first sensor and ends at the second sensor).
The parameter `positions` is a list of all the positions of the photoelectric sensors.
The parameter `times` is a list of the times when these sensors registered a passing object. That means, the first time corresponds to the first sensor, the second time to the second sensor, and so on.
In case a sensor didn't recognise a passing object, the `times` list contains `None` at the corresponding position. (In case you are wondering: missing times are inferred from the surrounding sensors.)

Try to calculate the average velocity of the object for each section!
If the calculation fails for a section (due to a missing end time), do the following: First, combine the section with the next section (for which then obviously the start time is missing).
Second, calculate the average velocity for the combined sections and use that velocity to estimate the time that the failed sensor would have measured.
Third, replace the `None` in the `times` list by the estimated value.
The function should return a tuple containing the updated `times` list and a list of the calculated velocities.

**Note:** Assume consistent units: either positions in m, times in s $\rightarrow$ velocities in m/s or positions in km, times in h $\rightarrow$ velocities in km/h.  
**Note:** A measurement series is completely discarded if the first or last sensors failed or two sensors in a row failed.
So you don't need to handle those special cases.

```python
# PROBLEM (2)

# SOLUTION
def velocity_calculator(positions, times):
    velocities = []
    for i in range(len(positions) - 1):
        delta_x = positions[i+1] - positions[i]
        try:
            delta_t = times[i+1] - times[i]
        except TypeError:
            delta_x_2 = positions[i+2] - positions[i]
            delta_t_2 = times[i+2] - times[i]
            velocity_2 = delta_x_2 / delta_t_2
            delta_t = delta_x / velocity_2
            times[i+1] = times[i] + delta_t
        velocities.append(delta_x / delta_t)
    return times, velocities

# PROBLEM-TEST
t1, v1 = velocity_calculator([0.0, 20.0, 50.0, 80.0, 100.0, 110.0, 125.0, 155.0, 170.0, 180.0, 200.0], [0.0, 10.0, None, 30.0, 35.0, 45.0, 55.0, None, 73.0, 80.0, 100.0])
t2, v2 = velocity_calculator([0.0, 13.0, 34.0, 45.0, 70.0, 88.0, 103.0, 142.0], [0.0, 18.0, None, 40.0, 55.0, None, 84.0, 90.0])
t3, v3 = velocity_calculator([142.0, 160.0, 185.6, 207.3, 223.0], [90.0, None, 108.8, None, 119.0])
for i in [t1, v1, t2, v2, t3, v3]:
    for j in range(len(i)):
        i[j] = round(i[j], 2)
(t1, v1), (t2, v2), (t3, v3)
```

Write a function called `triplet_sums` that takes two parameters as input: The first parameter is a list of integers called `container` and the scond parameter is a simple integer called `target_value`.

The function should find all triplets of numbers in the `container` that sum to the `target_value`.
The return value should be a list of all `sets` of three numbers that fulfill this requirement.

**Please Note**:
- The input list is guaranteed to not contain any duplicate numbers.
- Any number can only be used once per sum.
- Permutations are not allowed. Use sets to filter out permutations. For example, the output list should not contain both `{0, 1, 4}` and `{1, 4, 0}`.
- Illustrative Example: For `container = [1 ,2 ,3 ,4, 0, -1]` and `target_value = 5` the answer would be: `[{0, 1, 4}, {0, 2, 3}, {-1, 2, 4}]`
- If there is no solution, the returned list should be empty.

```python
# PROBLEM (2)
nums = [-2, 1,4, 2, -3, 3, 5, 6]
def triplet_sums(container, target_value):
# SOLUTION
    triplets = []
    n_numbers = len(container)

    # loop over all numbers, then looper over all numbers expect the one you already have, and then do the same again
    for i in range(n_numbers):
        for j in range(i+1, n_numbers):
            for k in range(j+1, n_numbers):
                if container[i] + container[j] + container[k] == target_value:
                    unique_combination = set((container[i], container[j], container[k]))
                    triplets.append(unique_combination)
    return triplets

# PROBLEM-TEST
passed = True
try:
    for value, target_length in zip([6, 10, 20, 1], [5, 3, 0, 3]):
        length_is_correct, sets_of_triplets_summing_to_value = False, True
        result = triplet_sums(nums, value)
#        print(result)
        length_is_correct = ( target_length == len(result))
        if target_length == 0:
            sets_of_triplets_summing_to_value = True
        else:
            for triplet in result:
                sets_of_triplets_summing_to_value = sets_of_triplets_summing_to_value and (len(triplet)==3) and isinstance(triplet, set) and (sum(triplet) == value)
#                print(sets_of_triplets_summing_to_value , (len(triplet)==3) , isinstance(triplet, set) , (sum(triplet) == value))
#        print(length_is_correct,sets_of_triplets_summing_to_value )
        passed = passed and length_is_correct and sets_of_triplets_summing_to_value
except:
    passed = False
# correct results
#([{-2, 2, 6}, {-2, 3, 5}, {1, 2, 3}, {-3, 4, 5}, {-3, 3, 6}],
# [{1, 4, 5}, {1, 3, 6}, {2, 3, 5}],
# [],
# [{-2, 1, 2}, {-3, -2, 6}, {-3, 1, 3}])

passed
```

### Problem 7 
You are working at a museum and your current task is to determine the materials of the latest discoveries of ~~tomb raider~~ archaeologist Mr. Jones. In front of you, you have 3 objects, each made from a different material. You determine their mass using a precise scale and measure their geometry using the fancy new laser measurement system of the museum:
- An ornate cube: Each side is 4 cm long and the mass is 0.51 kg.
- A polished sphere: The radius is 2 cm and the mass is 0.12 kg.
- A finely inscribed cylinder: Its height is 5 cm and its radius is 2 cm. Its mass is 0.63 kg.

Your goal is to identify the material of the objects based on their densities. A table (actually a `dictionary`) with likely material densities is given on the cell below. If the measured density lies within +- 5% of the table reference value, its very likely the material of the object.

**Task**:\
First calculate the volume of all objects.
Then add this information for each object to the `objects` dictionary under the key `"volume"`.
Furthermore, also include the mass information under the key `"mass"` for each repsective object.

In this task a clear distinction between object handling and material identification can be made. Write two functions called: `material_identifier` and `material_identification`. 
- `material_identification` function takes the `volume` and `mass` as arguments and determines the material from the given dictionary. The identified material is returned as a string. If the material cannot be identified, return a `"unknown"` string. 
- The `material_identifier` takes care of the object handling and takes as input a nested dictionary `objects_to_identify`. Within the function call the `material_identification` function for each object and include the material type under the key `"material"` for each object in the input dictionary. Since dictionarys are mutable objects, you can directly change the input dictionary, a return value is not necessary.

```python
# PROBLEM (2)
import math

density_table = {
    "polysterine": 20,  # kg/m^3,
    "cork": 240,        # kg/m^3
    "rhenium": 21020,   # kg/m^3
    "platinum": 21450,  # kg/m^3
    "copper": 8960,     # kg/m^3
    "iron": 7874,       # kg/m^3
    "diamond": 3515,    # kg/m^3
}

objects = {
    "cube" : {},
    "sphere" : {},
    "cylinder" : {},
}

# SOLUTION

cubic_side = 4e-2
objects["cube"]["volume"] = (cubic_side)**3  # m^3
objects["cube"]["mass"] = 0.51  # kg

sphere_radius = 2e-2
objects["sphere"]["volume"] = 4 / 3 * math.pi * (sphere_radius)**3  # m^3
objects["sphere"]["mass"] = 0.12  # kg

cylinder_height = 5e-2
cylinder_radius = 2e-2
objects["cylinder"]["volume"] = math.pi * (cylinder_radius)**2 * cylinder_height  # m^3
objects["cylinder"]["mass"] = 0.63  # kg

def material_identifier(objects_to_identify):
    for obj_key, obj_information in objects_to_identify.items():
        volume = obj_information["volume"]
        mass = obj_information["mass"]
        objects_to_identify[obj_key]["material"] = material_identification(volume, mass)

def material_identification(volume, mass):
    density = mass / volume  # kg/m^3

    factor_range = 0.05
    upper_bound = density * (1 + factor_range)
    lower_bound = density * (1 - factor_range)

    list_of_found_materials = []

    for material, material_density in density_table.items():
        if material_density >= lower_bound and material_density <= upper_bound:
            list_of_found_materials.append(material)
    # if the list contains exactly one material
    if len(list_of_found_materials)==1:
        return list_of_found_materials[0]
    # if none OR multiple are found
    return "unknown"

# PROBLEM-TEST

material_identifier(objects)
objects
```

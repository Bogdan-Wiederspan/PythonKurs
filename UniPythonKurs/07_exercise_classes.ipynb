### Problem 1

Create a class `Cylinder` that represents a solid cylinder with a mass m and radius r.
The `__init__` function needs to take the parameters `mass` and `radius` (with these names and in that order).
Implement a method `moment_of_inertia` (without arguments) that calculates and returns the moment of inertia of the cylinder with respect to its central axis.

````python
# PROBLEM (1)

# SOLUTION

class Cylinder:
    def __init__(self, mass, radius):
        self.mass = mass
        self.radius = radius

    def moment_of_inertia(self):
        return 0.5 * self.mass * self.radius**2

# PROBLEM-TEST
round(Cylinder(9, 0.2).moment_of_inertia(), 2), round(Cylinder(9, 0.2).moment_of_inertia(), 2), round(Cylinder(15, 0.5).moment_of_inertia(), 2)
````

### Problem 2

Write a class called `MyRectangle` which takes a height as first and a width as second parameter for its `__init__` function. Implement a method `get_height()` that returns the height and a method `get_width()` that returns the width.  Also implement a method called `area()` which returns its area, a method called `circumference()` which returns the circumference and a method called `diagonal()` which returns the length of the diagonal. 

*Hint: import the math module and use the `math.sqrt()` function*

````python
# PROBLEM (2)
import math  # for math.sqrt()

# SOLUTION

class MyRectangle:
    '''MyRectangle class'''
    def __init__(self, a=0, b=0):
        self.height = a
        self.width = b

    def get_height(self):
        return self.height

    def get_width(self):
        return self.width

    def area(self):
        return self.height * self.width

    def circumference(self):
        return 2 * self.height +  2 *self.width

    def diagonal(self):
        return math.sqrt(self.height * self.height + self.width * self.width)

# PROBLEM-TEST
rectangle = MyRectangle(3,7)
int(rectangle.get_height()), int(rectangle.get_width()),  int(rectangle.area()), int(rectangle.circumference()), round(rectangle.diagonal(),2)
````

### Problem 3

Write a class for three-dimensional vectors called `Vector3D`.

- Implement a method called `length()` that returns the length (norm) of the vector.
- Override the `__str__()` method to return a string of the form "(x, y, z)".
- Also implement a multiplication method for the multiplication of two vectors or the multiplication of a vector with a scalar. This method should overload the `*` operator.
    - If the vector is multiplied with another vector the result should be a scalar.
    - If the vector is multiplied with a scalar the return value should be a vector. Assume that the scalar is multiplied from the right: vector * scalar. (You do not need to care about the reverse operand order.)

*Hints:*
1) *To find out which method belongs to the multiplication `*` operator, you could check the documentation linked in the operator overloding section of the teaching notebook or create an integer object and check the output of the `dir()` function.*
2) *You can check the type of an object, for example the second operand, using the `isinstance()`function.*

````python
# PROBLEM (3)
import math

# SOLUTION

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def length(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

    def __str__(self):
        return f"({self.x}, {self.y}, {self.z})"

    def __mul__(self, otherObject):
        if isinstance(otherObject, Vector3D):
            return self.x * otherObject.x + self.y * otherObject.y + self.z * otherObject.z
        elif (isinstance(otherObject, float) or isinstance(otherObject, integer)):
            return Vector3D(otherObject * self.x, otherObject * self.y, otherObject * self.z)

# PROBLEM-TEST
# rounding to two digits should be failsafe enough
vector = Vector3D(3, 4, 5)
vector2 = Vector3D(2, 2, 2)
scalar = 5.0
vector.__str__().strip().replace(" ",""), round(vector.length(), 2), round((vector * scalar).length(), 2), int(vector * vector2)
````

````python
# PROBLEM (3)

# SOLUTION

class Particle:
    def __init__(self, mass, position, velocity):
        self.mass = mass
        self.position = position
        self.velocity = velocity

    def update_position(self, time_step):
        new_x = self.position[0] + self.velocity[0] * time_step
        new_y = self.position[1] + self.velocity[1] * time_step
        new_z = self.position[2] + self.velocity[2] * time_step
        self.position = (new_x, new_y, new_z)

    def kinetic_energy(self):
        speed_squared = self.velocity[0]**2 + self.velocity[1]**2 + self.velocity[2]**2
        return 0.5 * self.mass * speed_squared

    def get_x_position(self):
        return self.position[0]

    def get_mass(self):
        return self.mass


particles = [
    Particle(1, (0, 0, 0), (1, 2, 3)),
    Particle(2, (1, 2, 3), (-1, 2, -3)),
    Particle(3, (5, -2, 1), (0, 0, 1))
]

time_step = 2.5
for particle in particles:
    particle.update_position(time_step)

# Calculate the total kinetic energy of all particles
total_kinetic_energy = sum(particle.kinetic_energy() for particle in particles)

# Calculate the center of gravity in x of all particles
x_center_of_gravity = sum(particle.get_x_position() * particle.get_mass() for particle in particles) / sum(particle.get_mass() for particle in particles)

# PROBLEM-TEST
len(particles), round(particles[1].get_x_position(), 1), round(Particle(1, (32, 42, 52), (1, 1, 1)).get_x_position(), 1), round(total_kinetic_energy, 1), round(x_center_of_gravity, 1)
````

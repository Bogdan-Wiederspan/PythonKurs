### Problem 1

Write a function `even_squares(n)` that returns a list containing only the squares of all **even** natural numbers in the interval $[4, n]$ (both inclusive).

````python
# PROBLEM (1)

# SOLUTION
def even_squares(n):
    x = []
    for i in range(4, n + 1, 2):
        x.append(i**2)
    return x

# PROBLEM-TEST
tuple(map(sorted, (even_squares(100), even_squares(101), even_squares(4))))
````

### Problem 3

Write a function `remove_odd_values` with one parameter, which is a list that contains natural numbers up to 100 without duplicates (it is guaranteed that none of the entries is > 100 or appears twice in the list).
This function should remove all odd entries from the list and should return the list afterwards.
Make sure that the order of the remaining elements in the list stays the same.  
Solve the task without using `if` statements, which will be introduced in the next lesson.  
**Hint:** Use different data structures.

````python
# PROBLEM (1)

# SOLUTION
def remove_odd_values(x):
    odd_values = set(range(1, 100, 2))
    set_x = set(x)
    remove_values = set_x & odd_values
    for remove_value in remove_values:
        x.remove(remove_value)
    return x

# PROBLEM-TEST
remove_odd_values([4, 0, 7, 8, 5, 1, 22, 99, 100, 88, 35]), remove_odd_values([45, 100, 99, 1, 0, 7, 48, 83, 38, 76, 59, 94])
````

### Problem 5

A linear regression is the simplest fit one can do.
A linear function is given by the equation $y = m \cdot x + b$. 
We can solve the problem using the least squared method. 
The idea is to find the parameters `m` and `b` that minimize $\sum_i((m \cdot x_{i} + b) - y_{i})^2$ where $(x_i, y_i)$ are the measured points.

We can calculate the best fitting parameters using analystical methods as:
- $\displaystyle m = \frac{\sum_i((x_{i} - \bar{x})(y_{i} - \bar{y}))}{\sum_i(x_{i}-\bar{x})^2}$
- $\displaystyle b = \bar{y} - m \cdot \bar{x}$

where $\bar{x}$ and $\bar{y}$ are the means of $x$ and $y$, respectively.

Write a function `regression(x, y)` with two arguments `x`and `y`, where `x` is a list of measured $x$-values and `y` is a list of corresponding measured $y$-values.
The function should calculate $b$ and $m$ using the formulas given above.
The return statement should be a tuple that contains the two calculated parameters $m$ and $b$, in this order.

Use this function to calculate the Planck constant $h$ and work function $W_A$ of the **photoelectric effect**, which is described by the following equation:

$$E_\mathrm{kin} = h f - W_A$$

Save the results of the regression ($h$ and $- W_A$) in the variables `h` and `W`.  
Use the following data as input for the function `regression`:

The following table shows measurements of $E_\mathrm{kin}$ and $f$:
```
| (f / 10**15) [ Hz ] /   | 1.18 | 0.96 | 0.82 | 0.74 |
|-------------------------|------|------|------|------|
| E_kin / 10**(−19) [ J ] | 3.12 | 1.57 | 0.8  | 0.22 |
```
  
**Further reading:** If you want to read more about the photoelectric effect you can check out [this](https://www.leifiphysik.de/quantenphysik/quantenobjekt-photon/grundwissen/einsteins-theorie-des-lichts) article.
There is also a very similar exercise with solution available [here](https://www.leifiphysik.de/quantenphysik/quantenobjekt-photon/aufgabe/photoeffekt-abitur-2011-g8-ph12-a1-1)

**Hint:** The `zip` function might be very helpful for iterating over multiple lists and you might want to have a look at the [documentation](https://docs.python.org/3.3/library/functions.html#zip).


````python
# PROBLEM(2)

def regression(x,y):
# SOLUTION
    if len(x) != len(y):
        raise ValueError("x and y must have the same length")

    num_of_elements = len(x)

    x_mean = sum(x) / num_of_elements
    y_mean = sum(y) / num_of_elements

    # calculate m
    nominator = sum([(x_i - x_mean) * (y_i - y_mean) for x_i, y_i in zip(x,y)])
    denominator = sum([(x_i - x_mean)**2 for x_i in x])

    m = nominator / denominator
    # calculate b
    b = (y_mean - m * x_mean)
    return (m, b)


# PROBLEM-TEST
f = [1.18e15, 0.96e15, 0.82e15, 0.74e15]
E_kin = [3.12e-19, 1.57e-19, 0.8e-19, 0.22e-19]
h, W = regression(f, E_kin)
list(map(lambda x: (round(x[0], 3), round(x[1], 3)), (regression([0.9, 2.1, 2.8, 4.0, 5.2], [1.0, 2.9, 4.9, 7.1, 9.2]), regression([-6.1, -3.9, -2.2, 0.0, 2.1, 3.8, 6.2], [10.1, 8.9, 8.04, 6.98, 6.03, 5.01, 3.96]), regression([6, 4, 0, 2, 5, 8], [9, 4, 6, 2, 9, 1])))) + [(round(h, 37), round(W, 22))]
````

````python
# PROBLEM(2)

# SOLUTION
def sun_system_dict_creator(planets, distances_from_sun, masses, diameters):
    our_sun_system = {}
    for planet, distance, mass, diameter in zip(planets, distances_from_sun, masses, diameters):
        our_sun_system[planet] = {
            "distance_from_sun": distance,
            "mass": mass / 1e3,
            "radius": diameter / 2,
        }

    return our_sun_system


# PROBLEM-TEST
def precision_round(number, digits=3):
    power = "{:e}".format(number).split("e")[1]
    return round(number, -(int(power) - digits))

def round_sun_system_dict(our_sun_system):
    for key in our_sun_system:
        for key2 in our_sun_system[key]:
            our_sun_system[key][key2] = precision_round(our_sun_system[key][key2], 3)
    return our_sun_system

(
    round_sun_system_dict(
        sun_system_dict_creator(
            ("Sun", "Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune", "Pluto"),
            (0, 5.79e7, 1.08e8, 1.5e8, 2.28e8, 7.79e8, 1.43e9, 2.87e9, 4.49e9, 5.90e9),
            (1.99e33, 3.30e26, 4.87e27, 5.97e27, 6.42e26, 1.9e30, 5.68e29, 8.68e28, 1.02e29, 1.46e25),
            (1.392e6, 4.88e3, 1.21e4, 1.27e4, 6.79e3, 1.42e5, 1.20e5, 5.11e4, 4.95e4, 2.37e3),
        ),
    ),
    round_sun_system_dict(
        sun_system_dict_creator(
            ("Sun", "Milkyway", "Snickers", "Twix", "Earth", "Mars"),
            (0, 5.46e7, 1.47e8, 1.04e8, 2.17e8, 7.31e8, 1.96e9),
            (1.54e33, 3.79e26, 4.39e27, 5.78e27, 6.88e26, 1.52e30),
            (1.83e6, 4.47e3, 1.84e4, 1.48e4, 6.48e3, 1.11e5),
        ),
    ),
)
````

````python
# PROBLEM(2)
import math

pi = math.pi
G_constant = 6.673e-11  # (kg^−1 * m^3 * s^−2)

# SOLUTION
def planet_info_add(our_sun_system):
    for planet in our_sun_system:
        planet_information = our_sun_system[planet]

        planet_radius_in_meter =  planet_information["radius"] * 1e3  # now in m
        volume_planet = (4/3 * pi * planet_radius_in_meter**3)  # m^3
        density = planet_information["mass"] / volume_planet  # kg / m^3

        # calculate g
        g = G_constant * planet_information["mass"] / planet_radius_in_meter**2  # m/s^2
        # add information to dictionary
        our_sun_system[planet]["density"] = density
        our_sun_system[planet]["g"] = g
    return our_sun_system

# PROBLEM-TEST
def precision_round(number, digits=3):
    power = "{:e}".format(number).split("e")[1]
    return round(number, -(int(power) - digits))

def round_sun_system_dict(our_sun_system):
    for key in our_sun_system:
        for key2 in our_sun_system[key]:
            our_sun_system[key][key2] = precision_round(our_sun_system[key][key2], 3)
    return our_sun_system

(round_sun_system_dict(planet_info_add({'Sun': {'distance_from_sun': 0, 'mass': 1.99e+30, 'radius': 696000.0},
 'Mercury': {'distance_from_sun': 57900000.0,
  'mass': 3.2999999999999996e+23,
  'radius': 2440.0},
 'Venus': {'distance_from_sun': 108000000.0,
  'mass': 4.87e+24,
  'radius': 6050.0},
 'Earth': {'distance_from_sun': 150000000.0,
  'mass': 5.97e+24,
  'radius': 6350.0},
 'Mars': {'distance_from_sun': 228000000.0,
  'mass': 6.419999999999999e+23,
  'radius': 3395.0},
 'Jupiter': {'distance_from_sun': 779000000.0,
  'mass': 1.9e+27,
  'radius': 71000.0},
 'Saturn': {'distance_from_sun': 1430000000.0,
  'mass': 5.68e+26,
  'radius': 60000.0},
 'Uranus': {'distance_from_sun': 2870000000.0,
  'mass': 8.68e+25,
  'radius': 25550.0},
 'Neptune': {'distance_from_sun': 4490000000.0,
  'mass': 1.02e+26,
  'radius': 24750.0},
 'Pluto': {'distance_from_sun': 5900000000.0,
  'mass': 1.46e+22,
  'radius': 1185.0}})),
 round_sun_system_dict(planet_info_add({'Sun': {'distance_from_sun': 0,
  'mass': 1.5400000000000002e+30,
  'radius': 915000.0},
 'Milkyway': {'distance_from_sun': 54600000.0,
  'mass': 3.79e+23,
  'radius': 2235.0},
 'Snickers': {'distance_from_sun': 147000000.0,
  'mass': 4.39e+24,
  'radius': 9200.0},
 'Twix': {'distance_from_sun': 104000000.0,
  'mass': 5.78e+24,
  'radius': 7400.0},
 'Earth': {'distance_from_sun': 217000000.0,
  'mass': 6.88e+23,
  'radius': 3240.0},
 'Mars': {'distance_from_sun': 731000000.0,
  'mass': 1.5199999999999998e+27,
  'radius': 55500.0}})))
````

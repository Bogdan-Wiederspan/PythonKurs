- Revert the commit with the message `stupid commit`. Use `git revert` for this which is described [here](https://git-scm.com/docs/git-revert). If git is opening an editor and asks you to type a commit message, just save the suggested default message (**don't** change anything in the file). 
- Execute the `solution_code_generator.py` script on the current branch. Define a variable `code9` and assign it the code the script printed.

**Hint:** Many git commands are very powerful and offer (too) many options.
Often it is a good strategy to focus on the required arguments and ignore the other options.
In most cases the default does what you want.

**Good to know:** `git revert` creates a new commit to undo the changes of the commit you want to revert, it doesn't delete the commit, which means that the stupid commit is still in the history.
There are also options to delete commits but one should be very careful with those.
They change the history of your repository, which can cause serious problems if you work with many people and people have different histories in their repository.
Those commands should be only used for commits that are not yet published (pushed to any remote repo).
And still, be careful, if you change the history and you do something wrong, the original history is gone.

# Example problems

This demonstrates example problem types and their format such that they can be outograded. See the *_student_version* file for the auto-generated version the students will receive.

## Simple return value of the cell
Put the sample solution below `# SOLUTION`.
The last line defines the expected output which will be directly compared to the output of the student's solution.
This is fine for integers or strings (and floats for very simple cases -- not for results of multiple floating point operations).

**Example**

```python
# PROBLEM (1)
# SOLUTION
1 + 2 + 3
```

## Define a custom code to check the solution
If you want to check if, e.g., an implemented function does the right thing, you can define custom code that generates the output to be really checked below `# PROBLEM-TEST`. The students will not see the actual problem-test code. It will be appended to the cell when actually evaluating the notebooks for grading.  
You can add another `# SELF-CHECK` cell for the students to test their code to give them an idea what will be checked and to make sure they use the right signature (numer of parameters, no typos, ...). A self-check cell is not used for grading.

**Example**: For floating point results you should not test for equality. Rounding can be one solution.

Write a fuction `area_of_circle` taking exactly one parameter `r` that returns the area of a circle for the given radius

```python
# PROBLEM (1)
import math  # use math.pi in your solution

# SOLUTION
def area_of_circle(r):
    return math.pi * r**2

# PROBLEM-TEST
# rounding to two digits should be failsafe enough
round(area_of_circle(1), 2), round(area_of_circle(5), 2), round(area_of_circle(14), 2)
```

```python
# SELF-CHECK
# to test your solution, make sure this returns 1256.6370614359173
area_of_circle(20)
```

**Example**: Another option to check floating point results is to perform explicit comparisons with `math.isclose` to the expected results. They can be hardcoded, but also like this:

```python
# PROBLEM (1)
import math  # use math.pi in your solution

# SOLUTION
def area_of_circle(r):
    return math.pi * r**2

# PROBLEM-TEST
# rounding to two digits should be failsafe enough
import math
tuple(math.isclose(area_of_circle(r), math.pi * r**2, abs_tol=1e-2) for r in (1, 5, 14))
```

```python
# SELF-CHECK
# to test your solution, make sure this returns 1256.6370614359173
area_of_circle(20)
```

### Problem 2

Write a function `number_list_product` that calculates the product of all elements in a list. This function should have exactly one parameter which will always be a list of integer numbers. Return the product.

````python
# PROBLEM (1)

# SOLUTION
def number_list_product(list_of_numbers):
    prod = 1
    for element in list_of_numbers:
        prod *= element
    return prod

# PROBLEM-TEST
number_list_product([1, 2, 7, 9, -21, 101, 48, -67, 10, -60, 79, 24, 81, 3, 6, 18, -75, -66]), number_list_product([14, 57, 89, 98, -36, 29, -88, -14, 29, -73]), number_list_product([65, 25, -73, 84, -96, 56, 19, -81, 13])
````

### Problem 3

Write a function `sentence_list` that iterates over a list of sentence pieces (all of type string) and adds every piece to one string variable and returns this variable in the end. The function should have one parameter for the list of pieces.

````python
# PROBLEM (1)

# SOLUTION
def sentence_list(pieces):
    sentence = ""
    for piece in pieces:
        sentence += piece
    return sentence

# PROBLEM-TEST
pieces = ["I", " am", " a ", "comp", "lete sen", "tence", ",", " but I ", "onl", "y make s", "ense i", "f I am", " ", "conca", "tenated properly", ",", " which is your task."]
pieces2 = ["Hel", "lo Python ", "i", "s", " a ver", "y cool ", "progra", "ming la", "nguage!"]
pieces3 = ["I a", "m a demon", "strati", "on senten", "ce, only", " if I am", " cons", "tructe", "", "d cor", "rectly, ", "you w", "ill g", "et the po", "ints fo", "r this exer", "cise."]
sentence_list(pieces), sentence_list(pieces2), sentence_list(pieces3)
````

### Problem 4

Write a function `square_sum` that sums the squares of all numbers between 8 and `last_number` (inclusive on both sides) where `last_number` is a parameter of the function. Return the result.

````python
# PROBLEM (2)

# SOLUTION
def square_sum(last_number):
    result = 0
    for i in range(8, last_number + 1):
        result += i**2
    return result

# PROBLEM-TEST
int(square_sum(800)), int(square_sum(357)), int(square_sum(852))
````

### Problem 5

Write a function `sum_up_to` that sums all integer numbers starting from 1 and stopping once the sum exceeds `max_sum`, where `max_sum` is a parameter of the function. Return the sum.

````python
# PROBLEM (2)

# SOLUTION
def sum_up_to(max_sum):
    result = 0
    i = 1
    # iterate as long as max_sum is no stricly exceeded
    while result <= max_sum:
        result += i
        i += 1
    return result

# PROBLEM-TEST
sum_up_to(20000), sum_up_to(80), sum_up_to(15975378), sum_up_to(1236987456)
````

### Problem 6

Assume a system of $n$ point like bodies in one dimension, one of them has a mass $M$ of $10^{12}$ kg and is located at $x = 4000 m$. Write a function `gravitation(others)` that calculates the gravitational force on this body. The parameter `others` is a list that contains tuples with the mass and the position of the other bodies. Each tuple represents one other body and contains the mass at the first place in the tuple and the position of the body at the second place in the tuple. When you have such a tuple `tu` there are two ways to extract the mass and the position:  
```python
m, x = tu
```
or  
```python
m = tu[0]
x = tu[1]
```
You will learn more about tuples in the next lesson.  
Use Newton's law of gravitation  
$\displaystyle{F = G \frac{m_1 m_2 }{|r|^3} \cdot r}$  
to calculate the force between two bodies with $G$ being the gravitational constant (given below), $m_1$ and $m_2$ the masses of the two bodies and $r$ the distance between the two bodies. What is the total force on our selected body with mass $M$?

````python
# PROBLEM (3)
import math
G = 6.674e-11  # gravitational constant in m^3 kg^-1 s^-2

# SOLUTION
def gravitation(others):
    M = 1e12
    x = 4000
    F = 0
    for tu in others:
        m, y = tu
        r = y - x
        F += G * M * m * r / (abs(r**3))
    return F

# PROBLEM-TEST
round(gravitation([(8e11, 9000), (8e11, -1000)]), 2), round(gravitation([(1e13, 40000), (5e12, -10000), (6e12, -40000)]), 2), round(gravitation([(1e6, 9000), (5e7, -1000), (6e6, 5000)]), 2)
````
